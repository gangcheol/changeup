# 체인지업 기본 참여 마인드셋

`1` GDR-4 중 최합자 vs. 불합격자 비교

- 과제참여도는 전반적으로 높음

- 서류 제출 시기 늦음 : 미최합자의 `62%`는 4주차부터 본격적으로 서류 제출 시작
    * 완벽주의에 위험성...갖다 버리자

- 3C4P 분해 미흡 : 미최합자의 `50%`는 3C4P 완성도 미흡
    * **`3C4P`를 `매일 수치화`하고 `매일 디테일`하게 수정하자.**
    * 포트폴리오도 주차별로 업데이트 하자.(난 매일매일...)

- 중소기업 지원 비율 낮음 : 미최합자의 `50%`는 중소 기업의 지원률이 매우 저조하고 중견이상 지원
    * 중소기업을 쓰는 이유는 면접 연습이다.
    * 또한, 다 떨어질 경우를 대비해서 지원하는 것이 좋다.
- 지원 기업 준비도 미흡 : 미최합자의 `25%`는 각 회사별로 지원동기 작성을 못했거나 포트폴리오 또는 이력서를 자소서보다 우선시 하는 기업 위주로 지원

`2` 결론

- 1주차 부터 지원 시도하자.

- 한번을 하더라도 `영혼을 갈아넣어서` 완성주의 과제를 진행하자.

- 대기업만이 아닌 `대기업도 포함`해서 지원하자.

- 또한, 유튜브도 참고하자.(매일 매일 1개씩 보고 느낀점 기록)

# 경험리스트업으로 필살기 발굴

`1` 최종합격자들의 경험리스트업

- 신입은 분류한 총 경험개수 12개, 필살기 경험개수 3개 정도...

`2` 필살기 경험 유형

- 인문계열 경험유형 top3 : `인턴`, 계약직, 아르바이트

- 이공계열 경험유형 top3 : `학교프로젝트`, 연구/개발, 아르바이트

`3` 필살기/KPI 매칭 특징 : `수치화` 가능한 결과물이 있는 경험 `O`

- 인문계열 주요 KPI list : 매출상승(영업), 비용 및 예산준수(회계), 전환율(마케팅)

- 이공계열 주요 KPI list : 불량률 개선(품질관리), 성능개선(공정기술), 프로젝트 완료율(연구개발), 기술스택개선(IT개발)

`4` 최종합격자 경험리스트업케이스 $\to$ `나에게` 중요한 것이 아닌, `수치화가 된 결과물 위주로 경험 매칭!

- 데이터분석 : KPI 매칭이 1개만 될 경우도 있다. 그래도 걱정하지 말자
    * 연구개발
    * 수상경력
    * 연구개발
    * 교내 프로젝트
    * 피드백 : 내가 한 것들을 전부 수치화해서 다시 정리하자  

`6` 필살기 매칭 순위(취업전략 리포트를 참고해서 정리하자.)

`7` 경험리스트업 아쉬운 사례

- 유사경험을 점검해봐야할 경우 : 유사경험과 직무 KPI 매칭이 맞지 않는 경우
    * 연구개발하는 사람이 축구 1등 해놓고 성능개선한거다...뭐 이런식?

- KPI/경험 매칭이 어색한 경우
    * 내것도 뭔가 좀 어색하다. $\to$ 다시 첫 단추부터...ㅜㅜ
    * 직무 KPI부터 다시 엑셀로 뽑고 계속 갱신하자.
